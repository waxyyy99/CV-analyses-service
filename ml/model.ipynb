```python
from transformers import pipeline
import torch
# Простая настройка zero-shot
classifier = pipeline("zero-shot-classification", model="facebook/bart-large-mnli")

# Категории под твои критерии (адаптируй: 3 для структуры, 3 для оформления, 3 для наполнения)
structure_categories = ["плохая структура (нет разделов, хаос)", "средняя структура (частичные разделы, но не логично)", "хорошая структура (логичный порядок, все ключевые разделы)"]
formatting_categories = ["плохое оформление (длинные абзацы, нет ключевых слов)", "среднее оформление (читаемо, но без quantifiable метрик)", "хорошее оформление (короткие предложения, quantifiable достижения)"]
content_categories = ["плохое наполнение (мало навыков, пробелы в опыте)", "среднее наполнение (базовые навыки, но без деталей)", "хорошее наполнение (полный опыт, релевантные навыки)"]

print("Zero-shot готов! Теперь протестируем на примере резюме.")

# Пример текста резюме (вставь свой)
test_resume = """
John Doe
Software Engineer
Experience: Worked at Google for 2 years.
Skills: Python, Java.
Education: BS in CS from MIT.
"""

# Оценка по каждому критерию
def evaluate_resume_section(text, categories, section_name):
    result = classifier(text, candidate_labels=categories)
    top_label = result['labels'][0]
    top_score = result['scores'][0] * 10  # Нормализация в баллы 1–10
    return {section_name: {"label": top_label, "score": round(top_score, 2)}}

# Вызовы
structure_result = evaluate_resume_section(test_resume, structure_categories, "structure")
formatting_result = evaluate_resume_section(test_resume, formatting_categories, "formatting")
content_result = evaluate_resume_section(test_resume, content_categories, "content")

# Общая оценка
overall_score = (structure_result["structure"]["score"] + formatting_result["formatting"]["score"] + content_result["content"]["score"]) / 3

# Простые советы на основе результатов
advice = []
if structure_result["structure"]["score"] < 6:
    advice.append("Улучшите структуру: Добавьте четкие разделы 'Опыт', 'Навыки', 'Образование'.")
if formatting_result["formatting"]["score"] < 6:
    advice.append("Оформление: Используйте quantifiable метрики, e.g., 'Увеличил продажи на 20%' вместо 'Работал с продажами'.")
if content_result["content"]["score"] < 6:
    advice.append("Наполнение: Добавьте больше деталей в навыки и опыт, чтобы избежать пробелов.")

# Вывод
print("Результаты оценки:")
print(structure_result)
print(formatting_result)
print(content_result)
print(f"Общий балл: {round(overall_score, 2)}/10")
print("Советы по улучшению:")
for tip in advice:
    print(f"- {tip}")
```
